```python
def speed_loss_and_oee(file_name,output_count1,output_count2,speed):
    line_csv = pd.read_csv(file_name)
    line_csv_without_dash = line_csv.iloc[1:,:]
    dt3 = datetime.strptime('2021-10-20 06:00:48.00000', "%Y-%m-%d %H:%M:%S.%f")
    dt_line =line_csv_without_dash._get_value(1,0, takeable = True)
    line_time = datetime.strptime(dt_line, "%Y-%m-%d %H:%M:%S.%f")

    line_time_diff=round(((line_time-dt3).seconds)/60)
    output_count = float(line_csv_without_dash._get_value(1,output_count1, takeable = True))+ float(line_csv_without_dash._get_value(1,output_count2, takeable = True))
    print(output_count)
    if line_time_diff<480:
        present_time = (line_time_diff - 0)
        OEE = (output_count/(float(line_csv_without_dash._get_value(1,speed, takeable = True))*present_time))
        speed_loss = ((float(line_csv_without_dash._get_value(1,speed, takeable = True))*present_time)-output_count)/(float(line_csv_without_dash._get_value(1,speed, takeable = True))*present_time)
    if line_time_diff>480 and line_time_diff<960:
        present_time = line_time_diff - 480
        OEE = output_count/(float(line_csv_without_dash._get_value(1,6, takeable = True))*present_time)
        speed_loss = ((float(line_csv_without_dash._get_value(1,speed, takeable = True))*present_time)-output_count)/(float(line_csv_without_dash._get_value(1,speed, takeable = True))*present_time)


    if line_time_diff>960:
        present_time = line_time_diff - 960
        OEE = output_count/(float(line_csv_without_dash._get_value(1,6, takeable = True))*present_time)
        speed_loss = ((float(line_csv_without_dash._get_value(1,speed, takeable = True))*present_time)-output_count)/(float(line_csv_without_dash._get_value(1,speed, takeable = True))*present_time)

    print(OEE,speed_loss,(OEE+speed_loss))
```
